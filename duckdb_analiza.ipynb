# 📊 Analiza naslova vijesti pomoću DuckDB

```python
import duckdb
import pandas as pd
import time

# Učitavanje podataka
df = pd.read_csv("abcnews-date-text.csv")
df['publish_date'] = pd.to_datetime(df['publish_date'], format='%Y%m%d')

# Konektiranje na DuckDB u memoriji
con = duckdb.connect()
con.register("headlines_df", df)
```

```python
# Upit: Broj naslova po godini
start = time.time()
res = con.execute('''
    SELECT EXTRACT(YEAR FROM publish_date) AS godina, COUNT(*) AS broj
    FROM headlines_df
    GROUP BY godina
    ORDER BY godina
''').fetchdf()
end = time.time()
print(res)
print(f"Vrijeme: {end - start:.4f} sekundi")
```

```python
# Upit: naslovi koji sadrže riječ 'war'
start = time.time()
res2 = con.execute('''
    SELECT * FROM headlines_df
    WHERE headline_text ILIKE '%war%'
    LIMIT 5
''').fetchdf()
end = time.time()
print(res2)
print(f"Vrijeme: {end - start:.4f} sekundi")
```
